```python
class PVModule:
    def __init__(self, V_oc, I_sc, V_mp, I_mp):
        self.V_oc = V_oc
        self.I_sc = I_sc
        self.V_mp = V_mp
        self.I_mp = I_mp

    def calculate_output_voltage(self, irradiance, temperature):
        V = self.V_oc - (irradiance / 1000) * 0.1 - (temperature - 25) * 0.05
        return V


class MPPTController:
    def __init__(self, V_ref):
        self.V_ref = V_ref

    def track_voltage(self, pv_module, irradiance, temperature):
        V_out = pv_module.calculate_output_voltage(irradiance, temperature)
        error = self.V_ref - V_out
        return error


# Simulation parameters
V_ref = 36
irradiance = 1000
temperature = 25

# Create objects
pv_module = PVModule(V_oc=36, I_sc=8.3, V_mp=29, I_mp=7.6)
controller = MPPTController(V_ref)

# Run simulation
error = controller.track_voltage(pv_module, irradiance, temperature)
print("Error:", error)
```
